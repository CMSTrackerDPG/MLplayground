## Extracting variables from yaml file

```python
with open('./inputs/inputs_cfg.yaml') as file:
    inputs_dict = yaml.safe_load(file)
```

```python
for key, value in inputs_dict.items():
    print(key, value)
```

## Extracting variables from workspaces

```python
variables = [value[0] for key, value in inputs_dict.items()] # TO-DO: solve for groups with more than a single entry
```

```python
variables
```

```python
condition = df['path'].apply(lambda x: True if any(variable in x for variable in variables) else False)
df_variables = df[condition]
```
